```python
# Initialize Spark session
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("RDD Transformations and Actions").getOrCreate()
sc = spark.sparkContext

# Base RDD
rdd = sc.parallelize([1, 2, 3, 4, 5, 2, 3, 4])

mapped_rdd = rdd.map(lambda x: x * x)

filtered_rdd = rdd.filter(lambda x: x % 2 == 0)

flatmapped_rdd = rdd.flatMap(lambda x: range(1, x + 1))

distinct_rdd = rdd.distinct()

rdd2 = sc.parallelize([6, 7, 8])
union_rdd = rdd.union(rdd2)

intersection_rdd = rdd.intersection(rdd2)

subtract_rdd = rdd.subtract(rdd2)

cartesian_rdd = rdd.cartesian(rdd2)

sorted_rdd = rdd.sortBy(lambda x: x, ascending=False)

kv_rdd = rdd.map(lambda x: (x, 1))
reducedbykey_rdd = kv_rdd.reduceByKey(lambda a, b: a + b)  # word count style
groupedbykey_rdd = kv_rdd.groupByKey().mapValues(list)

print("Collect:", rdd.collect())

print("Count:", rdd.count())

print("First:", rdd.first())

print("Take 3:", rdd.take(3))

print("Top 3:", rdd.top(3))

print("Reduce (sum):", rdd.reduce(lambda a, b: a + b))

print("Sum:", rdd.sum())
print("Max:", rdd.max())
print("Min:", rdd.min())

print("CountByValue:", rdd.countByValue())

rdd.foreach(lambda x: print(x))


print("Mapped (x*x):", mapped_rdd.collect())
print("Filtered (even):", filtered_rdd.collect())
print("FlatMapped (range 1..x):", flatmapped_rdd.collect())
print("Distinct:", distinct_rdd.collect())
print("Union:", union_rdd.collect())
print("Intersection:", intersection_rdd.collect())
print("Subtract:", subtract_rdd.collect())
print("Cartesian (sample):", cartesian_rdd.take(5))
print("Sorted desc:", sorted_rdd.collect())
print("ReduceByKey (counts):", reducedbykey_rdd.collect())
print("GroupByKey:", groupedbykey_rdd.collect())

spark.stop()
```
